```python
#Exercise 2
def max_degree(Graph):
    mx = 0
    for node in Graph.nodes:
        if Graph.degree(node) > mx:
            result = (node,Graph.degree(node))
            mx = Graph.degree(node)
    return result

max_degree(Gr)
```
